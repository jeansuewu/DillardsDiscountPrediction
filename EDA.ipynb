```python
# join transact and skuinfo and strinfo
df = (pd
    .merge(trans, sku, left_on="SKU", right_on="SKU", how="left")
    .merge(store, left_on="STORE", right_on="STORE", how="left" ))


df.head()
```

```python
# calculate discount rate
df['discount_rate']=(df['AMT']-df['ORGPRICE'])/df['ORGPRICE']
df.head()
```

```python
# Extract month from sale date and add month column
d = df['SALEDATE'].astype(str).str.split("-")
month = [d[i][1] for i in range(len(df))]
df['month'] = month
df.month
```

```python
color_counts = df['COLOR'].value_counts()

# Define the criteria for encoding
def encode_color_popularity(count):
    if count > 10000:
        return 'Extreme'
    elif count > 5000:
        return 'Good'
    elif count > 1000:
        return 'Fair'
    else:
        return 'Low'

# Create a new column 'Color_Popularity' based on the criteria
df['Color_Popularity'] = df['COLOR'].map(color_counts.apply(encode_color_popularity))
```

```python
# One-Hot Encoding for month, state
df = pd.get_dummies(df, columns = ['month','STATE', 'Color_Popularity'])
df.head()
```
